```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as ani
def animatemyplot(i):
    mxdata=FileData[:,0][i+1]
    mydata=FileData[:,1][i+1]
    txdata=FileData[:,2][i+1]
    tydata=FileData[:,3][i+1]
    time=FileData[:,6][i+1]
    plot=plt.plot(mxdata,mydata, c='b', ls='-',marker='o')
    plot=plt.plot(txdata, tydata, c='k', ls='--', marker="^")
```

```python
%matplotlib notebook
filename='PG_NMT-K13.csv'
FileData = np.loadtxt(filename, skiprows=2, delimiter=',') #Load it in as an array
frames=len(FileData)
fig=plt.figure(figsize=(10,10))
plt.title(filename + " Animated Interception")
plt.xlabel('X Coordinate [m]')
plt.ylabel('Y Coordinate [m]')
animator=ani.FuncAnimation(fig, animatemyplot,frames,interval=10,repeat=True)
```
