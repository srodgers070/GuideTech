```python
def plotter(filename):
    import numpy as np
    import matplotlib.pyplot as plt
    updatefilename=filename+'.csv'
    data = np.loadtxt(updatefilename, skiprows=2, delimiter=',') #Load it in as an array

    print(len(data))
    plt.figure(1,figsize=(15,10))
    plt.subplot(311)
    plt.title("%s X vs Y data" %filename)
    plt.plot(data[:,0], data[:,1], ls='--', marker="o", label='Missile Positions')
    plt.plot(data[:,2], data[:,3], ls = '-', marker='^', label="Target Positions")
    plt.legend(loc='best')
    plt.xlabel("X Coordinate [m]")
    plt.ylabel("Y Coordinate [m]")
    plt.subplot(312)
    plt.plot(data[:,6],data[:,5], ls = "-", Label='Separation')
    plt.legend(loc='best')
    plt.ylabel('Separation [m]')
    plt.xlabel('Time [s]')
    plt.subplot(313)
    plt.plot(data[:,6],data[:,4], ls='-', label='Acceleration')
    plt.xlabel('Time [s]')
    plt.ylabel('Missile Acceleration [g]')
    plt.savefig(filename+"plot")
```
